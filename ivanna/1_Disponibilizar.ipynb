```python
#Importar librerías.
import pandas as pd
import requests
from bs4 import BeautifulSoup
import pickle
```

```python
#URL de TLC Trip Record Data (Datos de Viaje NY).
tlc_url = "https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page"
```

```python
#Solicitar URL y obtener objeto de respuesta.
tlc_respuesta = requests.get(tlc_url)
```

```python
#Analizar el texto obtenido.
tlc_sopa = BeautifulSoup(tlc_respuesta.text, 'html.parser')
```

```python
#Encontrar todos los hipervínculos presentes en la página web.
tlc_vinculos = tlc_sopa.find_all('a')
```

```python
#De todos los vínculos obtener los parquet para Yellow Taxi Trip Records.
i = 0
diccionario_amarillo = {}
años = ['2023', '2022', '2021', '2020', '2019', '2018']
for vinculo in tlc_vinculos:

    if (
        ('yellow' in vinculo.get('href', [])) and ('.parquet' in vinculo.get('href', []))
        ):

        for año in años:
        
            if (
                año in vinculo.get('href', [])
                ):
                i += 1
                print('Leyendo elemento: ', i)

                hipervinculo = vinculo.get('href', [])

                #Leer el vínculo y asignarlo a un diccionario.

                diccionario_amarillo["yellow_{0}_{1}".format(año, i)] = "yellow_{0}_{1} = pd.read_parquet('{2}')".format(año, i, hipervinculo)
                print("Elemento:", i, "escrito en diccionario.")
print("Todos los elementos se escribieron en el diccionario.")
```
