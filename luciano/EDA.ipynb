```python
Y = yellow[['trips_per_month','farebox_per_month','unique_vehicles','avg_hours_per_day_per_vehicle','avg_minutes_per_trip']]
corr_matrix = Y.corr(method='pearson')
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
G = green[['trips_per_month','farebox_per_month','unique_vehicles','avg_hours_per_day_per_vehicle','avg_minutes_per_trip']]
corr_matrix = G.corr(method='pearson')
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
F = FHVHV[['trips_per_month','unique_vehicles','avg_hours_per_day_per_vehicle','avg_minutes_per_trip']]
corr_matrix = F.corr(method='pearson')
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

## Trends

```python
Y['month_date'] = yellow['month_date']
G['month_date'] = green['month_date']
F['month_date'] = FHVHV['month_date']

```

```python
Y.nunique()
```

```python
# # Crear un rango de fechas que cubra los índices de todos los dataframes
# date_range = pd.date_range(start='2010-01-01', end='2023-04-07')
# # Establecer los mismos índices en cada dataframe
# Y = Y.reindex(date_range)
# G = G.reindex(date_range)
# F = F.reindex(date_range)
```

```python
fig, ax = plt.subplots()
sns.lineplot(x=Y.month_date, y=yellow['trips_per_day'], ax=ax, label='Yellow', color ='yellow')
sns.lineplot(x=G.month_date, y=green['trips_per_day'], ax=ax, label='Green', color='green')
sns.lineplot(x=F.month_date, y=FHVHV['trips_per_day'], ax=ax, label='FHVHV', color='blue')
ax.set_ylabel('Trips per day')
ax.set_title('Trips per day over time by type')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
sns.lineplot(x=Y.month_date, y=yellow['unique_vehicles'], ax=ax, label='Yellow',color ='yellow')
sns.lineplot(x=G.month_date, y=green['unique_vehicles'], ax=ax, label='Green',color ='green')
sns.lineplot(x=F.month_date, y=FHVHV['unique_vehicles'], ax=ax, label='FHVHV',color ='blue')
ax.set_ylabel('Unique_vehicles')
ax.set_title('Unique vehicles over time by type')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
sns.lineplot(x=Y.month_date, y=yellow['avg_hours_per_day_per_vehicle'], ax=ax, label='Yellow',color ='yellow')
sns.lineplot(x=G.month_date, y=green['avg_hours_per_day_per_vehicle'], ax=ax, label='Green',color ='green')
sns.lineplot(x=F.month_date, y=FHVHV['avg_hours_per_day_per_vehicle'], ax=ax, label='FHVHV',color ='blue')
ax.set_ylabel('avg_hours_per_day_per_vehicle')
ax.set_title('Hours per day over time by type')
ax.legend()
plt.show()
```

```python

fig, ax = plt.subplots()
sns.lineplot(x=Y.month_date, y=yellow['avg_minutes_per_trip'], ax=ax, label='Yellow',color ='yellow')
sns.lineplot(x=G.month_date, y=green['avg_minutes_per_trip'], ax=ax, label='Green',color ='green')
sns.lineplot(x=F.month_date, y=FHVHV['avg_minutes_per_trip'], ax=ax, label='FHVHV',color ='blue')
ax.set_ylabel('avg_minutes_per_trip')
ax.set_title('Minutes per trip over time by type')
ax.legend()
plt.show()
```
